```python
import pandas as pd
import plotly.express as px
```

```python
bull = pd.read_csv("bull_df.csv")
dates = list(bull['date'])
dates
```

```python
date_final = []
for i in range(0,len(dates)-1,5):
   date_final.append(dates[i])
```

```python
date_final = pd.DataFrame(date_final)
date_final.to_csv("bull_dates.csv")
```
