```python
app = Dash(external_stylesheets=[dbc.themes.CYBORG])
app.layout = html.Div([
    dbc.Row([
        dcc.DatePickerSingle(
            id='date1',
            min_date_allowed=date(2015, 8, 5),
            max_date_allowed=date(2022, 9, 19),
            initial_visible_month=date(2017, 8, 5),
            date=date(2017, 8, 25)),
        dcc.Graph(id= 'g', figure={})]
    )
])
@app.callback(
    Output('g', 'figure'),
    Input('date','val1')
)
def func(val1):
    if val1:
        d1 = datetime.strptime(val1, "%Y-%m-%d").strftime('%#d/%#m/%#Y')

        b = back_test_bull[back_test_bull['date1']>= d1]
        fig = px.line(b, x=b['date1'], y=b['Percent Change'])
        return fig
    else:
        Dash.no_update()
app.run_server(debug = False)
```
