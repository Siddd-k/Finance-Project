```python
ema = pd.read_csv('ema_bt.csv')
ema['date'] = pd.to_datetime(ema['date'])
ema['date1'] = ema['date'].dt.date
em = ema[ema['date1'] >= dtt.date(2022,1,1)]
em.head()
```

```python
#back testing for fortypoint strategy
def bt_ema(p):
    
    interest = ["Initial Capital"]
    date = ["-"]
    for i in em['Percent Change1']:
        interest.append(i)
    for i in em['date1']:
        date.append(i)
    #interest = back_test['Percent Change']
    capital = [p]
    for i in range(1,len(interest)):
        for j in range(len(capital)):
            if i  == (j + 1):
                t = (interest[i]*capital[j])/100 + capital[j]
                capital.append(t)
    a = pd.DataFrame({"date" : date,"Interest Rates" : interest,"Capital": capital})
    return a
```
